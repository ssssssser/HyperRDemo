```python
# Network X graph

import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Declare directed graph
G = nx.DiGraph()

nodes = ['Quality[p1]:0.7', 'Brand[p1]:Vaio', 'Category[p1]:Laptop', 'Sentiment[p1]:-0.95', 'Rating[p1]:2', 'Color[p1]:Silver', 'Price[p1]:999']
nodes += ['Quality[p2]:0.65', 'Price[p2]:529']

edges = [
    ('Quality[p1]:0.7','Price[p1]:999'), ('Quality[p1]:0.7','Rating[p1]:2'), ('Quality[p1]:0.7','Sentiment[p1]:-0.95'),
    ('Brand[p1]:Vaio','Quality[p1]:0.7'), ('Brand[p1]:Vaio','Price[p1]:999'), ('Brand[p1]:Vaio','Color[p1]:Silver'), ('Brand[p1]:Vaio','Rating[p1]:2'), ('Brand[p1]:Vaio','Sentiment[p1]:-0.95'), ('Brand[p1]:Vaio','Category[p1]:Laptop'),
    ('Category[p1]:Laptop','Sentiment[p1]:-0.95'), ('Category[p1]:Laptop','Rating[p1]:2'),
    ('Sentiment[p1]:-0.95','Rating[p1]:2'),
    ('Color[p1]:Silver','Rating[p1]:2'),
    ('Price[p1]:999','Rating[p1]:2'),
    
    ('Price[p2]:529','Rating[p1]:2'),
   ('Quality[p2]:0.65','Rating[p1]:2'),
]

# Add node or edge to networkx graph
print(nodes)
print(len(nodes))
for n in nodes:
    G.add_node(n)
for e in edges:
    source, target = e
    G.add_edge(source, target)
```
